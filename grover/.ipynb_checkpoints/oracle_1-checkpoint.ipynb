```python
# on-set = { 011 }, n_iteration = 2
# f = (000)'(001)'(010)'(100)'(101)'(110)'(111)'
from qiskit import *
from qiskit.quantum_info.operators import *
import numpy as np
%matplotlib inline

def MCT(qc, control_bits, target_bit, anci_bits):
    anci_idx = 0
    qc.ccx(control_bits[0], control_bits[1], anci_bits[anci_idx])
    for idx in range(2, len(control_bits) - 1):
        qc.ccx(control_bits[idx], anci_bits[anci_idx], anci_bits[anci_idx + 1])
        anci_idx += 1
    qc.ccx(control_bits[len(control_bits) - 1], anci_bits[anci_idx], target_bit)
    for idx in reversed(range(2, len(control_bits) - 1)):
        qc.ccx(control_bits[idx], anci_bits[anci_idx - 1], anci_bits[anci_idx])
        anci_idx -= 1
    qc.ccx(control_bits[0], control_bits[1], anci_bits[anci_idx])

cir = QuantumCircuit(16, 3) # 0~2: vars, 3~9: clauses, 10~14: mct ancillary, 15: ancillary
cir.x(0)
cir.x(1)
cir.x(2)
cir.h(0)
cir.h(1)
cir.h(2)

# 1st iteration
# oracle
cir.x(0)
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 3, range(10, 15)) # 000
cir.x(2)
MCT(cir, range(0, 3), 4, range(10, 15)) # 001
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 5, range(10, 15)) # 010
cir.x(0)
cir.x(1)
MCT(cir, range(0, 3), 6, range(10, 15)) # 100
cir.x(2)
MCT(cir, range(0, 3), 7, range(10, 15)) # 101
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 8, range(10, 15)) # 110
cir.x(2)
MCT(cir, range(0, 3), 9, range(10, 15)) # 111
cir.x(3)
cir.x(4)
cir.x(5)
cir.x(6)
cir.x(7)
cir.x(8)
cir.x(9)
MCT(cir, range(3, 10), 15, range(10, 15)) # 111
cir.z(15)
# inverse computation
MCT(cir, range(3, 10), 15, range(10, 15))
cir.x(3)
cir.x(4)
cir.x(5)
cir.x(6)
cir.x(7)
cir.x(8)
cir.x(9)
MCT(cir, range(0, 3), 9, range(10, 15))
cir.x(2)
MCT(cir, range(0, 3), 8, range(10, 15))
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 7, range(10, 15))
cir.x(2)
MCT(cir, range(0, 3), 6, range(10, 15))
cir.x(0)
cir.x(1)
MCT(cir, range(0, 3), 5, range(10, 15))
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 4, range(10, 15))
cir.x(2)
MCT(cir, range(0, 3), 3, range(10, 15))
cir.x(0)
cir.x(1)
cir.x(2)

# diffusion
cir.h(0)
cir.h(1)
cir.ccx(0, 1, 2)
cir.h(0)
cir.h(1)

# 2nd iteration
# oracle
cir.x(0)
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 3, range(10, 15)) # 000
cir.x(2)
MCT(cir, range(0, 3), 4, range(10, 15)) # 001
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 5, range(10, 15)) # 010
cir.x(0)
cir.x(1)
MCT(cir, range(0, 3), 6, range(10, 15)) # 100
cir.x(2)
MCT(cir, range(0, 3), 7, range(10, 15)) # 101
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 8, range(10, 15)) # 110
cir.x(2)
MCT(cir, range(0, 3), 9, range(10, 15)) # 111
cir.x(3)
cir.x(4)
cir.x(5)
cir.x(6)
cir.x(7)
cir.x(8)
cir.x(9)
MCT(cir, range(3, 10), 15, range(10, 15))
cir.z(15)
# inverse computation
MCT(cir, range(3, 10), 15, range(10, 15))
cir.x(3)
cir.x(4)
cir.x(5)
cir.x(6)
cir.x(7)
cir.x(8)
cir.x(9)
MCT(cir, range(0, 3), 9, range(10, 15))
cir.x(2)
MCT(cir, range(0, 3), 8, range(10, 15))
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 7, range(10, 15))
cir.x(2)
MCT(cir, range(0, 3), 6, range(10, 15))
cir.x(0)
cir.x(1)
MCT(cir, range(0, 3), 5, range(10, 15))
cir.x(1)
cir.x(2)
MCT(cir, range(0, 3), 4, range(10, 15))
cir.x(2)
MCT(cir, range(0, 3), 3, range(10, 15))
cir.x(0)
cir.x(1)
cir.x(2)

# diffusion
cir.h(0)
cir.h(1)
cir.ccx(0, 1, 2)
cir.h(0)
cir.h(1)

cir.barrier(range(0, 3))
cir.measure(range(0, 3), range(0, 3))
cir.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(cir, backend, shots = 16384)
result = job.result()

counts = result.get_counts(cir)
print(counts)
visualization.plot_histogram(counts)
```
